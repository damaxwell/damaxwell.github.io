You can make plots in Julia using the `Plots` package.  To use the package, you must first enter either `import Plots` or `using Plots`.  With `import Plots`, you must start all of the plotting commands with `Plots`, and with `using Plots` you can drop the `Plots` part.  We'll use `import Plots` for this question.

````julia
import Plots
````

To make a plot you make a vector with the $x$-coordinates, a vector with the $y$-coordinates.  Julia then connects the dots with straight lines.  If you've used MATLAB before, it works the same way.

Here's a basic example.


````julia
x = [0, 1, 2, 3]
y = [0.5, 2.8, 1.6, 4]

Plots.plot(x, y)
````

Here's how to make a graph of $y=\sin(x)$.  I'll show you first, and then explain the parts.

````julia
x = range(0, 2*pi, 100)
y = sin.(x)
Plots.plot( x, y, color=:green, linewidth=3, title="Sine", 
    xlabel="x", ylabel="y", legend = nothing)
````

Here are the steps:

1. ```julia
     x = range(0, 2*pi, 100 )
   ```
   
   
   Make a vector `x` that starts at 0, goes up to $2\pi$, and has 100 equally spaced entries.

2. ```julia
     y = sin.(x)
   ```

   Make a vector the same size as `x` but with entries `y[k] = sin(x[k])`.  Notice the period before the parenthesis.  This means "apply my function to each entry of the vector".


3. ```julia
      Plots.plot( x, y, color=:green, title="Sine", xlabel="x", ylabel="y", legend = nothing)   
   ```
    Now we plot the $x$-$y$ pairs. To make a nice plot, we've added some extras, like adding labels for the axes

In this problem, you are going to recreate the following figure:


There are eight stars on the unit circle with equal angles in between.  One of the stars is on the $x$-axis.

This is a scatter plot, where the $x$-$y$ coordinates have been specified just like for `Plots.plot`, but instead of connecting them with a line, markers are put down on the data points.

The command to make this is `Plots.scatter(x,y)`, and it takes in `x` and `y` just like for `Plots.plot`.  To get the nice appearance I used the following keyword arguments:
* `marker = :star` to get stars; the default is circles.
* `color = :yellow` (I'll let you guess what that did )
* `markersize = 8` to make the stars bigger than default
* `legend = nothing` to turn off the default legend
* `aspect_ratio=:equal` to make the $x$ and $y$ axes have the same aspect ratio so that the circle looks like a circle and not an oval.
