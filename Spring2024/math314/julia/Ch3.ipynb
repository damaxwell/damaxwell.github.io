```julia
import Plots
```

```julia
a = ones(10);
```

```julia
Plots.plot(a, color=:black)
Plots.scatter!(a,marker=:circle, color=:blue, markersize=8)
Plots.plot!(ylim=(-5,10), xlabel="x", ylabel="a_k", legend=nothing)
```

```julia
function plotdata(x,name)
    Plots.plot(x, color=:black)
    Plots.scatter!(x,marker=:circle, color=:blue, markersize=8)
    Plots.plot!(ylim=(-5,10), xlabel="k", legend=nothing)
    Plots.plot!(ylabel= name*"_k")
end
```

```julia
plotdata(a, "a")
```

```julia
b = 2.3 .+ 2.5*randn(10);
plotdata(b, "b")
```

```julia
c = 2.5 .+ 3.5*randn(10)
plotdata(c,"c")
```

```julia
d = 1.3 .+ 4.2*randn(10)
plotdata(d, "d")
```

```julia
function avg(x)
    return sum(x)/length(x)
end
function rms(x)
    return sqrt(x'*x / length(x))
end
function std(x)
    return rms(x .- avg(x))
end
```

```julia
risk = [std(a), std(b), std(c), std(d)]
ret = [avg(a), avg(b), avg(c), avg(d)]
labels = ("a", "b", "c", "d")
Plots.scatter(risk, ret, xlabel="risk", ylabel="return",
xlim=(-0.5,4.5), ylim=(0,3.5))
for k in 1:4
    Plots.annotate!(risk[k]+0.1, ret[k]+0.1, (labels[k],:bottom, :left ))
end
Plots.plot!(legend=nothing)
```
